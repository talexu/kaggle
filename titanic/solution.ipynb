```python
dataset = pd.get_dummies(df.drop(['Name', 'Ticket'], axis=1), columns=['PassengerId', 'Pclass', 'Sex', 'Cabin', 'Embarked'])
trainset = dataset.iloc[:700,]
valset = dataset.iloc[700:,]
dataset.head()
```
